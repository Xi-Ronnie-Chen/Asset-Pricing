# Module 1 Homework 2: Lognormal Returns

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import lognorm

%matplotlib inline
```

```python
def calc_mean(μ, T):
    return np.exp(μ*T)

def calc_std(μ, σ, T):
    return np.exp(μ*T) * np.sqrt(np.exp(σ**2*T) - 1)

def calc_sharpe(μ, σ, T):
    return (1 - np.exp(-μ*T)) / np.sqrt(np.exp(σ**2*T) - 1)
```

```python
μ = 0.06; σ = 0.20; T = 1

print(f'Mean:\t{100*(calc_mean(μ, T) - 1):.2f}')
print(f'Std:\t{100*calc_std(μ, σ, T):.2f}')
print(f'SR:\t{calc_sharpe(μ, σ, T):.2f}')
```

```python
μ = 0.06; σ = 0.20; T = 10

print(f'Mean:\t{100*(calc_mean(μ, T) - 1):.1f}')
print(f'Std:\t{100*calc_std(μ, σ, T):.1f}')
print(f'SR:\t{calc_sharpe(μ, σ, T):.1f}')
```

```python
μ = 0.06; σ = 0.20; T = 100

print(f'SR:\t{calc_sharpe(μ, σ, T):.4f}')
```

```python
Ts = np.linspace(1,50,50)
plt.plot(Ts, calc_mean(μ, Ts))
plt.xlabel('$T$'); plt.ylabel('Mean of T-Year Arithmetic Return')
pass
```

```python
Ts = np.linspace(1,50,50)
plt.plot(np.sqrt(Ts), calc_std(μ, σ, Ts))
plt.xlabel('$\sqrt{T}$'); plt.ylabel('Std of T-Year Arithmetic Return')
pass
```

```python
Ts = np.linspace(1,50,50)
plt.plot(np.sqrt(Ts), calc_sharpe(μ, σ, Ts))
plt.plot(np.sqrt(Ts), μ/σ * np.sqrt(Ts))
plt.xlabel('$\sqrt{T}$'); plt.ylabel('Sharpe Ratio')
pass
```

```python
μ = 0.06; σ = 0.20; T = 1

for T in [1, 5, 10, 50]:
    mean = (μ - 0.5*σ**2) * T
    std = σ * np.sqrt(T)
    print(f'T:\t{T}')
    print(f'Mean:\t{np.exp(mean + 0.5*std**2):.1f}')
    print(f'Median:\t{np.exp(mean):.1f}')
    print(f'Mode:\t{np.exp(mean - std**2):.1f}')
    print(f'Pr:\t{100*lognorm.cdf(calc_mean(μ, T), s=std, scale=np.exp(mean)):.1f}')
    print()
```
